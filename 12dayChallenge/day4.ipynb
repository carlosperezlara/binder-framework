3) How many different unique email domains names are there in the data set?

```python
df['emdomain'] = df['company email'].apply(lambda x:str(x).split('@')[-1])
len( df.groupby(['emdomain']) )
## NaN will be counted as one of the domains
```

4) Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
df_jobs = df.groupby(['NTA'])[ ["Current fulltime"] ]
df_bene = df.groupby(['NTA'])[ ["Savings from beginning receiving benefits"] ]

df_jobs_sum = df_jobs.apply(sum)
df_bene_sum = df_bene.apply(sum)
df_bene_cnt = df_bene.count()
df_bene_sum = df_bene_sum.divide( df_bene_cnt )

df_sum = pandas.concat( [df_jobs_sum, df_bene_sum], axis=1, join="inner" )
#df_sum.head()
```

```python
filt = df.groupby(['NTA'])['Company Name'].count()
filt = filt.mask( filt>=5 ).isna()
df_sum_NTA5 = df_sum[ filt ]
df_sum_NTA5.rename(columns={'Current fulltime':'JobsCreated','Savings from beginning receiving benefits':'AverageSavings'},inplace=True)
df_sum_NTA5
```
